```python
#Inserting into a tree
#BINARY SEARCH TREE

class Node:
    def __init__(self, data):
        
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        #Compare the new value with the parent node
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
            
#Print the tree
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print(self.data),
        if self.right:
            self.right.PrintTree()
            
#Use the insert methods to add nodes
root = Node(12)
root.insert(6)
root.insert(14)
root.insert(3)

root.PrintTree()
```

```python
#Binarytree library

from binarytree import tree, bst, heap

#Generate a random binary tree and return its root node
my_tree = tree(height = 4, is_perfect = False)

#Generate a random BST and return its root node
my_bst = bst(height = 4, is_perfect = True)

#Generate a random max heap and return its root node
my_heap = heap(height = 4, is_max = True, is_perfect = False)

#Pretty-print the trees in stdout
print(my_tree)
print(my_bst)
print(my_heap)
```

```python
#build own tree

from binarytree import Node

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print(root)

#Assert keyword - if condition returns True, nothing happens
assert root.height == 2
assert root.is_balanced is True
assert root.is_bst is False
assert root.is_complete is True
assert root.is_max_heap is False
assert root.is_min_heap is True
assert root.is_perfect is False #perfect - each internal node exactly 2 child
    #and all the leaves are at the same level
assert root.is_strict is True #strict - all its non-leaf nodes have both 
    #the left and right child node
assert root.leaf_count == 3
assert root.max_leaf_depth == 2
assert root.max_node_value == 5
assert root.min_leaf_depth == 1
assert root.min_node_value == 1
assert root.size == 5


#or all properties at once:
#assert root.properties == {.......}

print(root.leaves)
print(root.levels)
```

```python
#Compare and clone trees:

from binarytree import tree

original = tree()

#Clone the binary tree
clone = original.clone()

#Check if the trees are equal
original.equals(clone)
```

```python
#ude level-orded (breadth first) indexes to manipulate nodes

from binarytree import Node

root = Node(1)                 #index: 0 val: 1
root.left = Node(2)            #index: 1 val: 2
root.right = Node(3)           #index: 2 val: 3
root.left.right = Node(4)      #index: 4 val: 4
root.left.right.left = Node(5) #index: 9 val: 5

print(root)
root.pprint(index = True)
print(root[9])


#Replace the node/subtree at index 4
root[4] = Node(6, left = Node(7), right = Node(8))
root.pprint(index = True)

#Delete the node/subtree at index 1
del root[1]
root.pprint(index = True)
```

```python
#Traverse trees using different algorithms

from binarytree import Node

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print(root)

print(root.inorder)

print(root.preorder)

print(root.postorder)

print(root.levelorder)

print(list(root)) #the same as root.levelorder
```

```python
#Convert to list representations

from binarytree import build

#Build a tree from list representation
values = [7,3,2,6,9,None,1,5,8] #build according to levels
root = build(values)
print(root)

#go back to list representation
print(root.values)
```

```python
#another representation/more compact/but without indexing properties

from binarytree import build, build2, Node

#First let's create an example tree
root = Node(1)
root.left = Node(2)
root.left.left = Node(3)
root.left.left.left = Node(4)
root.left.left.right = Node(5)

print(root)

#FIRST representation - above
#All null nodes in each level are present
print(root.values)

#SECOND - more compact but without indexing
print(root.values2)

#Build trees from the list representations
tree1 = build(root.values)
tree2 = build2(root.values2)

assert tree1.equals(tree2) is True
```
